本代码爬取了当当网科普读物专栏的前10页，供600条信息。url如下：https://category.dangdang.com/pg1-cp01.52.04.01.00.00.html

```python
import requests
from bs4 import BeautifulSoup

# 定义一个函数来爬取指定的页面
def fetch_books(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
    
    # 发送请求并获取页面内容
    response = requests.get(url, headers=headers)
    
    # 如果请求成功，解析页面内容
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # 提取所有书籍信息的容器
        book_list = soup.find('ul', class_="bigimg").find_all('li')
        # print(book_list)

        # 遍历每个书籍，提取名称和价格
        books = []
        for book in book_list:

            # 提取书名
            book_name = book.find('a', class_='pic').get('title')
            
            # 提取价格
            price_tag = book.find('span', class_='search_now_price')
            if price_tag:
                price = price_tag.text.strip()  # 获取价格
            else:
                price = '价格未找到'  # 如果没有找到价格，则返回默认值

            # 提取作者
            try:
                author = book.find('p', class_="search_book_author").find('span').find('a', dd_name="单品作者").text.strip()
            except:
                author='佚名'
            # print(author)

            # 将书籍信息添加到列表中
            books.append({'book_name': book_name, 'price': price, 'author': author})
        
        return books
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return []

# 主程序
if __name__ == '__main__':
    for i in range(10): # 前十页内容
        url = 'https://category.dangdang.com/pg%s-cp01.52.04.01.00.00.html'%i  # 你可以根据需要修改页码
        books = fetch_books(url)
        
        # 打印爬取到的书籍信息
        for book in books:
            print(f"书名: {book['book_name']}, 价格: {book['price']}, 作者：{book['author']}")
```
